# Basic guidline for using data wrangler in Amazon Personalize

# Step1: Choose the dataset group that you create

# Step2: Choose the data set you are going to import

# Step3: Choose data wrangler as import method

keep clicking next will lead you to the console of Data Wrangler

# Step4: Import data to data wrangler

You can either choose amazon S3 or local file to import your data

# Step5: Start transform your data by “Add Step”

You can use "add step" for:  Change data type, filtering, drop missing column, drop column, 300+ built-in transformations and code snippets

All the step will also show in a data flow chart

# Step6: Export the data to S3 bucket
